```python
train['education'] = train['education'].replace(("Master's & above", "Bachelor's", "Below Secondary"),
                                                (3, 2, 1))
test['education'] = test['education'].replace(("Master's & above", "Bachelor's", "Below Secondary"),
                                                (3, 2, 1))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
train['department'] = le.fit_transform(train['department'])
test['department'] = le.fit_transform(test['department'])

train['gender'] = le.fit_transform(train['gender'])
test['gender'] = le.fit_transform(test['gender'])
```

```python
train.head()
```

```python
# Splitting the data
y = train['is_promoted']
x = train.drop(['is_promoted'], axis = 1)
x_test = test
```

```python
x.shape
```

```python
y.shape
```

```python
x_test.shape
```

```python
# handling Imblanced data

from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X, y = oversample.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split( X, y, test_size = 0.2, random_state = 0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_valid.shape)
print(y_valid.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_valid = sc.fit_transform(x_valid)
x_test = sc.transform(x_test)
```

```python

from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_valid)
```

```python
print("Traiinig Acc.: ", model.score(x_train, y_train))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_valid, y_pred)
sns.heatmap(cm, annot = True, cmap = 'Wistia')
plt.xlabel("Predictyed values")
plt.ylabel("Actual Values")
plt.show()
```

```python
cr = classification_report(y_valid, y_pred)
print(cr)
```
